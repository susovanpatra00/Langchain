## Text Summarization

We have many methods for text summarization. The methods are:
- Directly asking the LLM dor doing the task.   (Direct Method)
- Using Prompt Template for doing the task.     (Prompt Template Method)     
- Using different kinds of chain for doing the task. Different chains are:
    - Stuff Document Chain
    - Map Reduce Chain
    - Refine Chain

We can use any one of them, it does not matter in general. But if there is a case that 
the document size is greater than input token size of the LLM, then we can not use any 
random techniques. We mus use either `Map Reduce` or `Refine` technique.

* Stuff :   This is the simplest approach of combining documents. It concatenates all documents
            into a single string and passes that string to the language model in one go. This is
            useful when the size of the document is small enough to fit within the language model's
            context window.
* Map-Reduce :  It involves two steps: mapping(processing each document induvidually) and reducing
                (combining the processed results). Better than stuff as it solves the context length
                issue. But comuptationally rich. Best for large documents.
* Refine :  It is an iterative approach to document combination. It processes doucments one by one,
            and each processing step builds upon the previous one. Useful when order of the information
            processing matters. Slower than other two.

### Prompt Template

```python
from langchain.chains.llm import LLMChain
from langchain_core.prompts import PromptTemplate

template = '''
    Write a summary of the following text:
    Text: {text}
    Translate the precise summary into the following languages:
    Languages: {languages}
'''

prompt = PromptTemplate(
    input_variables= ['text', 'languages'],
    template= template
)

# Now I have the prompt, so need to create the chain to pass the input to the LLM.
```

```python
languages = ['Spanish', 'Bengali']

llm_chain = LLMChain(llm= llm, prompt= prompt)
summary2 = llm_chain.run({'text': text, 'languages': languages})
print(f"Summary: {summary2}")
```

### Map-Reduce Chain

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = PyPDFLoader("Attention.pdf")
docs = loader.load_and_split(text_splitter= RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=100))
docs
```

```python
# I will use the whole pdf here.
len(docs)
```

```python
chunk_prompt = '''
                    Summarize the below text:
                    Text: {text}
                    Summary:
                '''

map_prompt_template = PromptTemplate(
    input_variables=['text'],
    template= chunk_prompt
)

final_prompt =  '''
                    Give final summary of the entire document. 
                    Add a Introduction and Conclusion page.
                    Text: {text}
                '''
final_prompt_template = PromptTemplate(
    input_variables=['text'],
    template= final_prompt
)
```

```python
summarize_chain = load_summarize_chain(
    llm= llm, 
    chain_type= 'map_reduce',
    map_prompt = map_prompt_template,
    combine_prompt = final_prompt_template, 
    verbose= True
)

summary = summarize_chain.run(docs)
summary
```

### Refine 

```python
# Question Prompt
initial_prompt_template = PromptTemplate(
    input_variables=["text"],
    template="""
                Write a concise summary of the following text:
                Text: {text}
            """
)

# Refine Prompt
refine_prompt_template = PromptTemplate(
    input_variables=["existing_answer", "text"],
    template="""
                We have provided an existing summary: {existing_answer}
                We would like you to refine this summary based on the following text:
                Text: {text}
                If the existing summary is sufficient, do not add anything.
            """
)



refine_chain = load_summarize_chain(
    llm= llm,
    chain_type= 'refine',
    question_prompt=initial_prompt_template,
    refine_prompt=refine_prompt_template,
    verbose= True,
)

summary = refine_chain.run(docs)
summary
```
